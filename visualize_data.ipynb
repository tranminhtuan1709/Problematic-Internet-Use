## Train.csv

```python
df_train = pd.read_csv(
    "Dataset/child-mind-institute-problematic-internet-use/train.csv"
)
```

```python
df_train.to_excel(
    excel_writer="Output/train.xlsx", header=True, index=True, engine="openpyxl"
)
```

```python
all_features = df_train.columns.to_list()[1:-1]
print("All features:", all_features)
print("Total:", len(all_features), "features")
```

### Insight of Baisc_Desmos features

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.countplot(x="Basic_Demos-Enroll_Season", data=df_train, ax=axes[0])
sns.kdeplot(x="Basic_Demos-Age", data=df_train, ax=axes[1], fill=True)
sns.countplot(x="Basic_Demos-Sex", data=df_train, ax=axes[2])

fig.suptitle("Distribution of Basic Demographic Features", y=1.05)

plt.tight_layout()
plt.show()
```

### Insight of CGAS features

```python
df_train["CGAS-CGAS_Score"].max()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 5))

sns.countplot(data=df_train, x="CGAS-Season", ax=axes[0])
sns.kdeplot(data=df_train, x="CGAS-CGAS_Score", ax=axes[1], fill=True)

fig.suptitle("Distributions of CGAS Features", y=1.05)

plt.tight_layout()
plt.show()
```

### Insight of Physical features

```python
fig, axes = plt.subplots(2, 4, figsize=(15, 5))

sns.countplot(data=df_train, x="Physical-Season", ax=axes[0][0])
sns.kdeplot(data=df_train, x="Physical-BMI", ax=axes[0][1], fill=True)
sns.kdeplot(data=df_train, x="Physical-Height", ax=axes[0][2], fill=True)
sns.kdeplot(data=df_train, x="Physical-Weight", ax=axes[0][3], fill=True)
sns.kdeplot(data=df_train, x="Physical-Waist_Circumference", ax=axes[1][0], fill=True)
sns.kdeplot(data=df_train, x="Physical-Diastolic_BP", ax=axes[1][1], fill=True)
sns.kdeplot(data=df_train, x="Physical-HeartRate", ax=axes[1][2], fill=True)
sns.kdeplot(data=df_train, x="Physical-Systolic_BP", ax=axes[1][3], fill=True)

fig.suptitle("Distributions of Physical features", y=1.05)

plt.tight_layout()
plt.show()
```

### Insight of Fitness features

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

sns.countplot(data=df_train, x="Fitness_Endurance-Season", ax=axes[0][0])
sns.kdeplot(data=df_train, x="Fitness_Endurance-Time_Mins", ax=axes[0][1], fill=True)
sns.kdeplot(data=df_train, x="Fitness_Endurance-Time_Mins", ax=axes[1][0], fill=True)
sns.kdeplot(data=df_train, x="Fitness_Endurance-Time_Sec", ax=axes[1][1], fill=True)

fig.suptitle("Distributions of Fitness features", y=1.05)

plt.tight_layout()
plt.show()
```

### Insight of FGC features

```python
fig, axes = plt.subplots(5, 3, figsize=(10, 10))

sns.countplot(data=df_train, x="FGC-Season", ax=axes[0][0])
sns.histplot(data=df_train, x="FGC-FGC_CU", ax=axes[0][1], kde=True)
sns.histplot(data=df_train, x="FGC-FGC_CU_Zone", ax=axes[0][2], kde=True)
sns.kdeplot(data=df_train, x="FGC-FGC_GSND", ax=axes[1][0], fill=True)
sns.histplot(data=df_train, x="FGC-FGC_GSND_Zone", ax=axes[1][1], kde=True)
sns.kdeplot(data=df_train, x="FGC-FGC_GSD", ax=axes[1][2], fill=True)
sns.histplot(data=df_train, x="FGC-FGC_GSD_Zone", ax=axes[2][0], kde=True)
sns.histplot(data=df_train, x="FGC-FGC_PU", ax=axes[2][1], kde=True)
sns.histplot(data=df_train, x="FGC-FGC_PU_Zone", ax=axes[2][2], kde=True)
sns.kdeplot(data=df_train, x="FGC-FGC_SRL", ax=axes[3][0], fill=True)
sns.histplot(data=df_train, x="FGC-FGC_SRL_Zone", ax=axes[3][1], kde=True)
sns.kdeplot(data=df_train, x="FGC-FGC_SRR", ax=axes[3][2], fill=True)
sns.histplot(data=df_train, x="FGC-FGC_SRR_Zone", ax=axes[4][0], kde=True)
sns.kdeplot(data=df_train, x="FGC-FGC_TL", ax=axes[4][1], fill=True)
sns.kdeplot(data=df_train, x="FGC-FGC_TL_Zone", ax=axes[4][2], fill=True)

fig.suptitle("Distributions of FGC features", y=1.05)

plt.tight_layout()
plt.show()
```
