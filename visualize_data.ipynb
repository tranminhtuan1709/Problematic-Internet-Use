## Analysis of features by SII Target Distribution

### Continuous features

```python
continuous_features = [
    "Basic_Demos-Age",
    "CGAS-CGAS_Score",
    "Physical-BMI",
    "Physical-Height",
    "Physical-Weight",
    "Physical-Waist_Circumference",
    "Physical-Diastolic_BP",
    "Physical-HeartRate",
    "Physical-Systolic_BP",
    "Fitness_Endurance-Max_Stage",
    "Fitness_Endurance-Time_Mins",
    "Fitness_Endurance-Time_Sec",
    "FGC-FGC_CU",
    "FGC-FGC_GSND",
    "FGC-FGC_GSD",
    "FGC-FGC_PU",
    "FGC-FGC_SRL",
    "FGC-FGC_SRR",
    "FGC-FGC_TL",
    "BIA-BIA_BMC",
    "BIA-BIA_BMI",
    "BIA-BIA_BMR",
    "BIA-BIA_DEE",
    "BIA-BIA_ECW",
    "BIA-BIA_FFM",
    "BIA-BIA_FFMI",
    "BIA-BIA_FMI",
    "BIA-BIA_Fat",
    "BIA-BIA_ICW",
    "BIA-BIA_LDM",
    "BIA-BIA_LST",
    "BIA-BIA_SMM",
    "BIA-BIA_TBW",
    "PAQ_A-PAQ_A_Total",
    "PAQ_C-PAQ_C_Total",
    "PCIAT-PCIAT_Total",
    "SDS-SDS_Total_Raw",
    "SDS-SDS_Total_T",
]

fig, axes = plt.subplots(10, 4, figsize=(12, 20))

row = 0
col = 0

for feature in continuous_features:
    if col >= 4:
        row += 1
        col = 0

    sns.kdeplot(
        data=df_train,
        x=feature,
        ax=axes[row][col],
        fill=True,
        hue="sii",
        palette="Set2",
        legend=False,
    )

    col += 1

for row in range(10):
    for col in range(4):
        if col != 0:
            axes[row][col].set_ylabel("")

sii_labels = sorted(df_train["sii"].dropna().unique())
colors = sns.color_palette("Set2", len(sii_labels))
legend_elements = [
    plt.scatter([], [], color=color, lw=4, label=f"SII = {label}")
    for color, label in zip(colors, sii_labels)
]

fig.legend(
    handles=legend_elements,
    loc="upper center",
    ncol=4,
    bbox_to_anchor=(0.5, 1.15),
    frameon=False,
)

axes[9][2].set_visible(False)
axes[9][3].set_visible(False)

fig.suptitle("Distributions of continuous features by SII target distribution", y=1.05)

plt.tight_layout()
plt.show()
```

## Feature statistics

```python
all_features = df_data_dictionary["Field"].to_list()[1:]
str_features = df_data_dictionary[df_data_dictionary["Type"].str.contains("str")][
    "Field"
].to_list()[1:]

numeric_features = [feature for feature in all_features if feature not in str_features]

feature_statistics = {
    "Feature": [],
    "Null percentage": [],
    "Unique value": [],
    "Min": [],
    "Max": [],
    "Mean": [],
    "Variance": [],
    "Median": [],
}

df_feature_statistics = pd.DataFrame(feature_statistics)

for feature in numeric_features:
    new_row = {
        "Feature": feature,
        "Null percentage": df_train[feature].isna().mean() * 100,
        "Unique value": df_train[feature].nunique(),
        "Min": df_train[feature].min(),
        "Max": df_train[feature].max(),
        "Mean": df_train[feature].mean(),
        "Variance": df_train[feature].var(),
        "Median": df_train[feature].median(),
    }

    df_feature_statistics.loc[len(df_feature_statistics)] = new_row

df_feature_statistics

df_feature_statistics.to_excel(
    excel_writer="Output/feature_statistics.xlsx",
    header=True,
    index=True,
    engine="openpyxl",
)
```
